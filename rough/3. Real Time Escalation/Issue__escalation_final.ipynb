```python
import gspread
import pandas as pd
from collections import Counter

# Install gspread if not already installed
!pip install "gspread==6.1.3"

# Replace with your actual sheet link and API key
Sheet_link="https://docs.google.com/spreadsheets/d/1QBhwbKvbEaWZUuY6RpHLGw-VpcEAJvoTMBjMr-3dol0/edit?usp=sharing"
key = "AIzaSyBcb9DOkzDzWiFQJcdDxPrV-ChYCtdOucY" # Replace with your actual API key

gc = gspread.api_key(key)

sh = gc.open_by_key("1QBhwbKvbEaWZUuY6RpHLGw-VpcEAJvoTMBjMr-3dol0")
worksheet = sh.worksheet("Sheet1")
data = worksheet.get_all_values()
df = pd.DataFrame(data[1:], columns=data[0])

df_en = df.loc[df["language"].eq("en"), :]

# Very high priority will be assigned to the issues having critical tags or body content
def should_escalate(incoming_issue):
    if incoming_issue["priority"] == "high":
        # Check both tags and body for keywords
        tags_combined = " ".join([str(incoming_issue.get(f"tag_{i+1}", "")) for i in range(0, 9)])
        body = str(incoming_issue.get("body", ""))  # Assuming the body column exists

        # Check for keywords that escalate to 'VERY_HIGH' in both tags and body
        very_high_keywords = [
            "Disruption", "Failure", "Outage", "Crash", "Breach", "Critical", "Down", "Unavailable", "Malfunction", "Not working",
            "Hack", "Vulnerability", "Compromise", "Attack", "Exploited", "Leak", "Insecurity", "Emergency", "Incident",
            "Catastrophic", "Severe", "Shutdown", "System down", "Data loss", "Corruption", "Compliance failure", "Violation",
            "Risk", "Breach of policy", "Legal issue", "Degraded", "Severe lag", "Unresponsive"
        ]
        for key in very_high_keywords:
            if key.lower() in tags_combined.lower() or key.lower() in body.lower():
                return "VERY_HIGH"

        # Check for additional keywords that escalate to 'HIGH' in both tags and body
        high_keywords = [
            "Delay", "Lag", "Slow", "Queue", "Timeout", "Performance issue", "Issue", "Problem", "Concern", "Complication",
            "Bug", "Error", "Failure to connect", "Non-functional", "Interruption", "Risk", "Security issue", "Privacy concern",
            "Unauthorized access", "Access failure", "Inconvenience", "Frustrating", "Not responsive", "Not accessible",
            "Disruptive", "User problem", "Partial outage", "Problematic", "Minor issue", "Limited functionality", "Misconfiguration"
        ]
        for key in high_keywords:
            if key.lower() in tags_combined.lower() or key.lower() in body.lower():
                return "HIGH"

    # Return the original priority if no escalation
    return incoming_issue["priority"]

df_en['escalation_priority'] = df_en.apply(should_escalate, axis=1)

# Display relevant columns
print(df_en[['subject', 'escalation_priority']].head(30))
```
